# Serialize and store the private key securely

```python
def store_private_key(private_key, password):
    salt, encryption_key = derive_key(password)

    private_key_bytes = private_key.private_bytes(
        encoding=serialization.Encoding.PEM,
        format=serialization.PrivateFormat.PKCS8,
        encryption_algorithm=serialization.BestAvailableEncryption(encryption_key)
    )
    return salt
```


# Derive a symmetric encryption key from the private key

```python
def derive_key(password):
    salt = os.urandom(16)
    kdf = PBKDF2HMAC(
        algorithm=hashes.SHA256(),
        iterations=100000,
        salt=salt,
        length=MAX_KEY_SIZE
    )
    key = kdf.derive(password)
    return salt, key

password = b"your_super_secret_password"
salt, encryption_key = derive_key(password)
```


# Encrypt data

```python
def encrypt_data(data, key, iv):
    cipher = Cipher(algorithms.AES(key), modes.CFB(iv))
    encryptor = cipher.encryptor()
    padder = padding.PKCS7(algorithms.AES.block_size).padder()

    padded_data = padder.update(data) + padder.finalize()
    ciphertext = encryptor.update(padded_data) + encryptor.finalize()
    return ciphertext

plaintext_data = b"Hello IoT World!"
iv = os.urandom(MAX_IV_SIZE)
ciphertext = encrypt_data(plaintext_data, encryption_key, iv)
```

# Decrypt data

```python
def decrypt_data(ciphertext, key, iv):
    cipher = Cipher(algorithms.AES(key), modes.CFB(iv))
    decryptor = cipher.decryptor()
    unpadder = padding.PKCS7(algorithms.AES.block_size).unpadder()

    decrypted_padded_data = decryptor.update(ciphertext) + decryptor.finalize()
    decrypted_data = unpadder.update(decrypted_padded_data) + unpadder.finalize()
    return decrypted_data

decrypted_data = decrypt_data(ciphertext, encryption_key, iv)
print("Decrypted Data:", decrypted_data.decode('utf-8'))
```

##Measuring execution times for encryption and decryption key scheduling

```python
import time
import matplotlib.pyplot as plt

# Define the number of iterations for each process
NUM_ITERATIONS = 1000

# Lists to store execution times
key_schedule_times = []
encryption_times = []
decryption_times = []

# Measure execution time for key scheduling
for _ in range(NUM_ITERATIONS):
    start_time = time.time()

    # Perform key scheduling process here

    end_time = time.time()
    key_schedule_times.append(end_time - start_time)

# Measure execution time for encryption
for _ in range(NUM_ITERATIONS):
    start_time = time.time()

    # Perform encryption process here

    end_time = time.time()
    encryption_times.append(end_time - start_time)

# Measure execution time for decryption
for _ in range(NUM_ITERATIONS):
    start_time = time.time()

    # Perform decryption process here

    end_time = time.time()
    decryption_times.append(end_time - start_time)

# Plotting the results
plt.plot(range(NUM_ITERATIONS), key_schedule_times, label='Key Scheduling')
plt.plot(range(NUM_ITERATIONS), encryption_times, label='Encryption')
plt.plot(range(NUM_ITERATIONS), decryption_times, label='Decryption')
plt.xlabel('Iteration')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time Comparison: Key Scheduling vs. Encryption vs. Decryption')
plt.legend()
plt.show()
```
